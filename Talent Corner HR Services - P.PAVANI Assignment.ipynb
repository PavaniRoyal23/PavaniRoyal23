```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate synthetic stock data as an example
dates = pd.date_range(start='2020-01-01', end='2023-01-01', freq='B')
np.random.seed(42)
```

```python
dates
```

```python
# Simulate adjusted closing prices with some volatility
price_changes = np.random.randn(len(dates)) * 2  # Random walk
prices = 100 + np.cumsum(price_changes)  # Cumulative sum to simulate price movement
prices
```

```python
# Create a DataFrame
stock_data = pd.DataFrame({'Date': dates, 'Adjusted Close': prices})
stock_data.set_index('Date', inplace=True)
stock_data
```

```python
# Calculate moving averages to understand trends over different time periods
stock_data['MA_30'] = stock_data['Adjusted Close'].rolling(window=30).mean()  # 30-day moving average
stock_data['MA_90'] = stock_data['Adjusted Close'].rolling(window=90).mean()  # 90-day moving average
```

```python
stock_data['MA_30']
```

```python
stock_data['MA_90']
```

```python
# Calculate daily returns to analyze volatility
stock_data['Daily Return'] = stock_data['Adjusted Close'].pct_change()
stock_data['Daily Return']
```

```python
# Display the recreated data with moving averages and daily returns for continuity check
stock_data.head()
```

### The dataset is successfully recreated, now containing:

Adjusted Close prices,

30-day (MA_30) and 90-day (MA_90) moving averages for trend analysis,

Daily Return for volatility assessment.

```python
# Plotting Adjusted Close with Moving Averages
plt.figure(figsize=(12, 6))
plt.plot(stock_data.index, stock_data['Adjusted Close'], label='Adjusted Close', color='blue', linewidth=1.5)
plt.plot(stock_data.index, stock_data['MA_30'], label='30-Day Moving Average', color='orange', linestyle='--')
plt.plot(stock_data.index, stock_data['MA_90'], label='90-Day Moving Average', color='green', linestyle='--')
plt.title('Stock Price with 30 and 90-Day Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting Daily Return Distribution to assess volatility
plt.figure(figsize=(8, 6))
sns.histplot(stock_data['Daily Return'].dropna(), kde=True, color='purple', bins=30)
plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Summarizing daily return statistics for insights on volatility
daily_return_stats = stock_data['Daily Return'].describe()
daily_return_stats
```

### 2.Marketing Strategy Based on Al Insights

##### Let's begin with trend forecasting on the stock price data using a time series model. We’ll use a simple yet effective model, ARIMA (Auto-Regressive Integrated Moving Average), which is well-suited for financial data with time dependencies.

Data Preparation: Split the data into training and test sets.
    
ARIMA Model: Train an ARIMA model on the training data.
    
Forecasting: Use the model to forecast future stock prices.
    
Visualization: Plot actual vs. forecasted values to assess the model’s accuracy.

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
# Split data into training and testing sets (80% train, 20% test)
train_size = int(len(stock_data) * 0.8)
train_data, test_data = stock_data['Adjusted Close'][:train_size], stock_data['Adjusted Close'][train_size:]
```

```python
train_size
```

```python
train_data, test_data
```

```python
# Fit ARIMA model on training data (ARIMA parameters chosen here as a starting example)
model = ARIMA(train_data, order=(5, 1, 0))  # (p,d,q) order where d=1 for differencing to make data stationary
arima_model = model.fit()
```

```python
model
```

```python
arima_model
```

```python
# Forecast for the length of the test data
forecasted_values = arima_model.forecast(steps=len(test_data))
forecasted_values
```

```python
# Plotting actual vs. forecasted values
plt.figure(figsize=(12, 6))
plt.plot(train_data.index, train_data, label='Training Data', color='blue')
plt.plot(test_data.index, test_data, label='Actual Prices', color='green')
plt.plot(test_data.index, forecasted_values, label='Forecasted Prices', color='red', linestyle='--')
plt.title('Stock Price Forecasting using ARIMA')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 1. Predictive Analytics for Customer Churn Prediction

Dataset: You can use the Telco Customer Churn dataset from Kaggle.
    
Tools: Use scikit-learn for model building (logistic regression or decision trees).
    
Steps:
Load the dataset and preprocess it (handling null values, encoding categorical variables).

Use logistic regression to predict if a customer will churn.

Evaluate the model with accuracy and F1 scores.

```python
import pandas as pd

# Load the Telco Customer Churn dataset
telco_data = pd.read_csv('Telco Customer Churn.csv')

# Display basic information and first few rows of the dataset to understand its structure
telco_data_info = telco_data.info()
telco_data_head = telco_data.head()

telco_data_info, telco_data_head
```

#### The Telco Customer Churn dataset has 7,043 entries and 21 columns, including customer demographic information, account details, service usage, and churn status.

```python
from sklearn.preprocessing import LabelEncoder

# Convert 'TotalCharges' to numeric, handling any non-numeric values (e.g., empty strings)
telco_data['TotalCharges'] = pd.to_numeric(telco_data['TotalCharges'], errors='coerce')
telco_data = telco_data.dropna(subset=['TotalCharges'])  # Drop rows with null TotalCharges values

# Encode categorical columns using LabelEncoder
label_enc = LabelEncoder()
for column in telco_data.select_dtypes(include=['object']).columns:
    if column != 'customerID':  # Exclude customerID from encoding
        telco_data[column] = label_enc.fit_transform(telco_data[column])

# Display the first few rows to verify the preprocessing steps
telco_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Define features (X) and target (y)
X = telco_data.drop(['customerID', 'Churn'], axis=1)
y = telco_data['Churn']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the logistic regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Make predictions and evaluate the model
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

accuracy, classification_rep
```

##### The logistic regression model achieved an accuracy of approximately 79.7% in predicting customer churn

Precision and Recall: For customers who did not churn (class 0), the model has a high precision (84%) and recall (89%). For customers who did churn (class 1), precision is 64% and recall is 53%.
F1-Score: The F1-score is 0.87 for non-churned customers and 0.58 for churned customers, indicating the model is more accurate in predicting customers who stay than those who leave.

Customer Retention Focus: The model can help identify customers at higher risk of churning, allowing targeted retention efforts.
    
Predictive Value: The model provides valuable insights into the features contributing to churn, which can be further refined for better accuracy.
